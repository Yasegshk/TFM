```python
#Creo un dataframe temporal sólo con los datos numéricos
df_copy = df

for i in df.columns:
    if str(df_copy[i].dtype) == 'object':
        df_copy.drop([i], axis = 'columns', inplace=True)

df_copy
```

Se observa que las únicas columnas con valores nulos son tres y corresponden con aquellas asociadas a la muerte del paciente. Por lo tanto, estos valores nulos implica que dicho paciente no ha fallecido, por lo que se podrían reemplazar los NaN por un cero (por ejemplo)

```python
df['days_to_death'] = df['days_to_death'].fillna(0)
df['year_of_death'] = df['year_of_death'].fillna(0)

'''
Otras columnas con nulos que no se como tratar:
year_of_birth, ajcc_pathologic_stage, ajcc_staging_system_edition, days_to_last_follow_up, year_of_diagnosis
'''
```

Otro aspecto importante junto con los valores nulos es el tratamiento de los posibles outliers.

```python
df_copy_norm=(df_copy-df_copy.mean())/df_copy.std()
df_copy_norm.boxplot()
```

Aunque se observan muchos outliers, esperemos ver cuales son las pruebas que realizaremos y así no comprometamos todas las variables. De igual forma, de desecharlos comenzaría por los que están por encima de 4 y por debajo de -4 (en esta escala normalizada).

```python
# HISTOGRAMA
#Se establece el tamaño de las imagenes
fig, axs = plt.subplots(32,2)
fig.set_size_inches(100, 100)
x=-1
#Para cada columna, se representa un histograma con el valor de un registro y el número de veces que se repite
for col in df_copy.columns:
    x+=1
    axs[x//2,x%2].hist(df_copy[col],bins=25)
    axs[x//2,x%2].set_title('%s' %col)
for ax in axs.flat:
    ax.set(ylabel = 'Número de registros')

#Código extraído de la asignatura de minería de datos
```

A continuación se lleva a cabo un test de normalidad. Para poder rechazar la hipótesis nula, los datos deben estar por debajo de 0.05. Vemos que en una gran mayoría de las columnas no puede ser rechazada.

```python
#Test de normalidad
from scipy import stats
scores=[]
for name in df_copy.columns:
  _, p_value = stats.normaltest(df_copy[name])
  print(name,' P-valor = ',p_value)
```

```python
#El siguiente paso es comprobar la correlación entre los datos

import itertools
import seaborn as sns
from numpy import corrcoef, transpose, arange
from pylab import pcolor, show, colorbar, xticks, yticks
```

```python
names=df_copy.columns #Guardamos los nombres de las columnas
train_T=df_copy.transpose()
```

```python
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.corrcoef.html
R = corrcoef(train_T)

# http://stanford.edu/~mwaskom/software/seaborn/examples/many_pairwise_correlations.html
# Generate a mask for the upper triangle
sns.set(style="white")
mask = np.zeros_like(R, dtype=bool)
mask[np.triu_indices_from(mask,k=1)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(60, 60))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(200, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(R, mask=mask, cmap=cmap, vmax=1, vmin=-1, annot=True,fmt=".2f",
            square=True, xticklabels=names, yticklabels=names,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)

'''
Problemas con year_of_birdth
'''
```
