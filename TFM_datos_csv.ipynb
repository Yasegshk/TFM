A continuaci√≥n, mediante el siguiente bucle for, recorro todos los archivos txt de la carpeta, elimino columnas inecesarias, calculo la media por columna y guardo el array resultante en un nuevo dataframe.

```python
import pandas as pd
import numpy as np


pd.options.display.max_columns = None

inicio = pd.read_csv("c:/Users/Yasa/Desktop/TFM/histomicstk/TXT_caracteristicas/" + nombres_archivos[0], sep='\t', engine='python')
columnas = list(inicio.columns)
DATOS = pd.DataFrame(columns=columnas, index=range(145)) #Hay que definir el tama√±o del df para poder usar la funcion iloc[]
DATOS.drop(['Object ID', 'Centroid X ¬µm', 'Centroid Y ¬µm', 'Name', 'Class', 'Parent', 'ROI'], axis = 'columns', inplace=True)

for i in range(len(nombres_archivos)):
    data = pd.read_csv("c:/Users/Yasa/Desktop/TFM/histomicstk/TXT_caracteristicas/" + nombres_archivos[i], sep='\t', engine='python')
    data.drop(['Object ID', 'Centroid X ¬µm', 'Centroid Y ¬µm', 'Name', 'Class', 'Parent', 'ROI'], axis = 'columns', inplace=True)
    columnas = list(data.columns)

    medias = list()
    a = data['Image'][0].split('.')
    b = a[0] + a[1]
    medias.append(b)

    for j in columnas[1:]:
        x = data[j].mean()
        medias.append(round(x, 4))
    DATOS.iloc[i] = medias
DATOS
```

El siguiente paso es unir cada fila de datos con su correspondiente ID del paciente y subtipo de c√°ncer. Para ello nos ayudamos del archivo Subtype.csv

```python
subtipos = pd.read_csv("C:/Users/Yasa/Desktop/TFM/Subtype.csv")
subtipos
```

Para facilitar el proceso, creo de antemano las columnas necesarias en nuestro dataframe de datos y las relleno con numeros del 1 al 145 (numero de filas). Estos n√∫meros ser√°n posteriormente reemplazados con los datos correspondientes a cada fila.

```python
DATOS['Study ID'] = range(145)
DATOS['Patient ID'] = range(145)
DATOS['Subtype'] = range(145)
```

```python
for i in range(145):
    imagen = DATOS['Image'][i]
    for j in range(len(subtipos)):
        ID = subtipos['Patient ID'][j]
        if imagen[:12] == ID:
            DATOS['Study ID'][i] = subtipos['Study ID'][j]
            DATOS['Patient ID'][i] = subtipos['Patient ID'][j]
            DATOS['Subtype'][i] = subtipos['Subtype'][j]
```
